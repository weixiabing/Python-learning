```python
#加载txt
import numpy as np
import matplotlib.pyplot as plt
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

f=open('prices.txt','r')
#f转换为两列
data=np.loadtxt(f,delimiter=',')
f.close()
```

```python
#计算每列平均值、方差，直方图
print(data.shape)
print(data.mean(axis=0))
print(data.var(axis=0))
plt.hist(data[:,0],bins=100)
plt.title('第一列直方图')
plt.show()
plt.hist(data[:,1],bins=100)
plt.title('第二列直方图')
plt.show()

```

```python
#对每列进行z-score标准化  
data_zscore=(data-data.mean(axis=0))/data.std(axis=0)  
print(data_zscore.mean(axis=0))  
print(data_zscore.std(axis=0))  
plt.hist(data_zscore[:,0],bins=100)  
plt.show()  
#保存标准化后的数据  
np.savetxt('prices_zscore.txt',data_zscore,delimiter=',') 
```
